```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
from tensorflow.keras.callbacks import TensorBoard
import time
name="cat-dogs-cnn-64x2-{}".format(int(time.time()))
tensorboard=TensorBoard(log_dir="C:\\logs\\{}".format(name))
```

```python
DATADIR="C:\\Users\\LENOVO!!\\Downloads\\PetImages"
CATEGORIES = ["Cat","Dog"]

for i in CATEGORIES:
    path=os.path.join(DATADIR,i) # GETTING THE PATH 
    for img in os.listdir(path): # ITERATING OVER EACH IMAGE
        image=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(image, cmap='gray')
        plt.show()
        break
print(image.shape)        
```

```python
IMG_SIZE = 50

new_array = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap='gray')
plt.show()           
```

```python
training_data=[]
def create_training_data():
    for i in CATEGORIES:
        path=os.path.join(DATADIR,i) # GETTING THE PATH 
        num=CATEGORIES.index(i)
        
        for img in tqdm(os.listdir(path)): # ITERATING OVER EACH IMAGE
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,num])
            except Exception as e:    
                pass
create_training_data()
```

```python
import random
random.shuffle(training_data)
for i in training_data[:23]:
    print(i[1])
```

```python
# DIVIDE DATA INTO FEATURES AND LABELS
X=[]
Y=[]
for features, labels in training_data:
    X.append(features)
    Y.append(labels)
X=np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,1)
print(X.shape)
```

```python
import pickle
f=open("X.pickle","wb")
pickle.dump(X,f)
f.close()

f=open("Y.pickle","wb")
pickle.dump(Y,f)
f.close()

```

```python
import pickle
import numpy as np
X=np.asarray(pickle.load(open("X.pickle","rb")))
Y=np.asarray(pickle.load(open("Y.pickle","rb")))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Activation, Flatten,Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential

X=X/255.0
model=Sequential()

model.add(Conv2D(64,(3,3), input_shape=X.shape[1:]))

model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dense(1))

model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X,Y,batch_size=32,epochs=3,validation_split=0.3,callbacks=[tensorboard])
model.summary()
```
